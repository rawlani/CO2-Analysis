# Implementing the analysis using Pandas library

```python
import pandas as pd
```

Opening up the relevant files and saving with relevant names

```python
file_hawaii = 'C:\\Users\\HP\\co2_hawaii.txt'
file_alaska = 'C:\\Users\\HP\\co2_alaska.txt'
```

Selecting columns which are required for analysis purposes

```python
col_name = ['site_code', 'year', 'month', 'day', 'hour', 'minute', 'second', 'value', 'value_unc', 'nvalue', 'latitude', 'longitude', 'altitude', 'elevation', 'intake_height', 'instrument', 'qcflag']
```

Created a dataframe with name 'data_hawaii' and read the contents of files to it. 

I have shown only top 5 of the contents of list so to make sure what kind of data the file has

```python
data_hawaii = pd.read_table(file_hawaii, comment='#', delim_whitespace=True, names=col_name)
data_hawaii.head()
```

Created a dataframe with name 'data_alaska' and read the contents of files to it.

```python
data_alaska = pd.read_table(file_alaska, comment='#', delim_whitespace=True, names=col_name)
data_alaska.head()
```

Parsed the Quality Control Flags of Hawaii data by keeping only those rows in the dataset which are of our use in final calculations

```python
new_data_hawaii = pd.DataFrame(data_hawaii[data_hawaii['qcflag'] != '*..']).reset_index()
new_data_hawaii.head()
```

Parsed the Quality Control Flags of Alaska data by keeping only those rows in the dataset which are of our use in final calculations

```python
new_data_alaska = pd.DataFrame(data_alaska[data_alaska['qcflag'] != '*..']).reset_index()
new_data_alaska.head()
```

Extracting Maximum and Average data per year for Hawaii Countrt

```python
h_max = pd.DataFrame(new_data_hawaii.groupby('year')[['value']].max())
h_mean = pd.DataFrame(new_data_hawaii.groupby('year')[['value']].mean())
```

```python
h_max.head()
```

```python
h_mean.head()
```

Extracting Maximum and Average data per year for Alaska country

```python
a_max = pd.DataFrame(new_data_alaska.groupby('year')[['value']].max())
a_mean = pd.DataFrame(new_data_alaska.groupby('year')[['value']].mean())
```

```python
a_max.head()
```

```python
a_mean.head()
```

Merging the two dataframe which were just created in above steps

```python
final_hawaii_data = pd.merge(h_max, h_mean, on='year')
```

```python
final_hawaii_data.columns = ['MAX_LEVEL', 'MEAN_LEVEL']
```

Below is the calculation for measuring the percent Change in the mean level compared to previous years for Hawaii Country

```python
final_hawaii_data['% CHANGE'] = (((final_hawaii_data['MEAN_LEVEL']-(final_hawaii_data['MEAN_LEVEL'].shift(1)))/final_hawaii_data['MEAN_LEVEL'])*100).fillna(0)
```

```python
final_hawaii_data['Country'] = 'Hawaii'
```

```python
final_hawaii_data.reset_index().head()
```

```python
final_alaska_data = pd.merge(a_max, a_mean, on='year')
```

```python
final_alaska_data.columns = ['MAX_LEVEL', 'MEAN_LEVEL']
```

Below is the calculation for measuring the percent Change in the mean level compared to previous years for Alaska Country

```python
final_alaska_data['% CHANGE'] = (((final_alaska_data['MEAN_LEVEL']-(final_alaska_data['MEAN_LEVEL'].shift(1)))/final_alaska_data['MEAN_LEVEL'])*100).fillna(0)
```

```python
final_alaska_data['Country'] = 'Alaska'
```

```python
final_alaska_data.reset_index().head()
```

```python
annual_co2 = pd.concat([final_alaska_data, final_hawaii_data])
```

```python
annual_co2.reset_index()
```
